```python
class Vehicle:
    
    def __init__(self,char, axis ,x, y,length,fuel):
        if char.isalpha():
            self.char = char
        else:
            raise ValueError('Not a letter.')
        
        self.axis = axis
        
#         self.x = x
#         self.y = y
        
        if 0 <= x <=5:
            self.x = x
        else:
              raise ValueError('x is wrong size')
        if 0 <= y <=5:
            self.y = y
        else:
             raise ValueError('y is wrong size')
       
        self.length = length
            
        self.fuel = int(fuel)
        
    
    def set_fuel(val):
        self.fuel = val
    
    def set_axis(val):
        self.axis = val
    def set_x(val):
        self.x = val
    def set_y(val):
        self.y = val
    def set_length(val):
        self.length = val
        
    def __hash__(self):
        return hash((self.char,self.x,self.y))

    def __eq__(self, other):
        return self.char == other.char and self.x == other.x and self.y == other.y

    def __ne__(self, other):
        return not self.__eq__(other)

    def __repr__(self):
        return "Vehicle({0}, {1}, {2}, {3})".format(self.char, self.x, self.y,
                                                    self.axis)
    
    
            
    def __str__(self):
        return f"Vehicle({self.char},{self.axis},{self.x},{self.y},{self.length},{self.fuel})"
```

```python
def string_to_matrix(line,n):
    temp = [line[char: char + n] for char in range(0, len(line), n)]
    
    matrix = [list(x) for x in temp]
    
    return np.array(matrix)
    
```

```python
def matrix_vehicles(line,n):
    cars0_fuels = line.split()
    matrix = string_to_matrix(cars0_fuels[0],n)
    fuels = {}
    temp = []
    encountered = set()
    for pair in cars0_fuels[1:]:
        fuels[pair[0]] = pair[1]
    for car in cars0_fuels[0]:
        if car not in fuels:
            fuels[car] = 100
    
    for row_count,row in enumerate(matrix):
        for col,ele in enumerate(row):
            if ele == "." or ele in encountered:
                continue
            else:
                if col == len(row)-1: #end of the row / last col in row
                    if matrix[row_count][col-1] == ele: #check if left is the same value
                        axis = "horizontal"
                    else:
                        axis = "vertical"
                elif col == 0: #beginning of a row /first col in row
                    if matrix[row_count][col+1] == ele:  #check if right is the same value
                        axis = "horizontal"
                    else:
                        axis = "vertical"
                else: # somewhere in row
                    if matrix[row_count][col+1] == ele or matrix[row_count][col-1]== ele: 
                        axis = "horizontal"
                    else:
                        axis = "vertical"
                        
                encountered.add(ele)
                length = cars0_fuels[0].count(ele)
                car_fuel = fuels.get(ele)
                new_car = Vehicle(ele,axis,row_count,col,length,car_fuel)
                temp.append(new_car)
    return temp
                
        
    
```

```python
class Board(object):
    
    def __init__(self,cars1,change):
        self.cars = cars1
        self.change = change
        

    def __hash__(self):
        s = '-' * 8 + '\n'
        for line in self.get_board():
            s += '|{0}|\n'.format(''.join(line))
        s += '-' * 8 + '\n'
        return hash(s)

    def __eq__(self, other):
        return self.cars == other.cars

    def __ne__(self, other):
        return not self.__eq__(other)
    
    def __repr__(self):
        return self.change
    
    
    def get_change(self):
        return(self.change)
    
    def get_board(self):
        emp = np.full([6, 6],'.', dtype="str")
        for stuff in self.cars:
            if stuff.axis == 'vertical':
                emp[stuff.x:stuff.x+stuff.length,stuff.y] = stuff.char
            if stuff.axis == 'horizontal':
                #pprint(vars(stuff))
                emp[stuff.x,stuff.y:stuff.y+stuff.length] = stuff.char
                
        list_of_lists = list()
        for row in emp:
            list_of_lists.append(row.tolist())
                    
        return list_of_lists
    
        
    
    def moves(self):
        """Return iterator of next possible moves."""
        board = self.get_board()
        for v in self.cars:
            if v.fuel <= 0: 
                continue
            if v.axis == 'vertical':
                if v.x - 1 >= 0 and board[v.x - 1][v.y ] == '.':
                    #print(f"will subtract 1 from {v.x}")
                    new_v = Vehicle(v.char,v.axis, v.x - 1, v.y,v.length,v.fuel-1)
                    new_vehicles = self.cars.copy()
                    new_vehicles.remove(v)
                    new_vehicles.add(new_v)
                    yield Board(new_vehicles,f'{v.char} up')
                if v.x + v.length <= 5 and board[v.x + v.length ][v.y] == '.':
                   # print(f"will add 1 to {v.x}")
                    new_v = Vehicle(v.char,v.axis, v.x + 1, v.y,v.length,v.fuel-1)
                    new_vehicles = self.cars.copy()
                    new_vehicles.remove(v)
                    new_vehicles.add(new_v)
                    yield Board(new_vehicles,f'{v.char} down')
            else:
                if v.y - 1 >= 0 and board[v.x][v.y - 1] == '.':
                    #print(f"will subtract 1 from {v.y}")
                    new_v = Vehicle(v.char,v.axis, v.x, v.y - 1 ,v.length,v.fuel-1)
                    new_vehicles = self.cars.copy()
                    new_vehicles.remove(v)
                    new_vehicles.add(new_v)
                    yield Board(new_vehicles,f'{v.char} left')
                if v.y + v.length <= 5 and board[v.x ][v.y+ v.length] == '.':
                    #print(f"will add 1 to {v.y}")
                    new_v = Vehicle(v.char,v.axis, v.x, v.y +1 ,v.length,v.fuel-1)
                    new_vehicles = self.cars.copy()
                    new_vehicles.remove(v)
                    new_vehicles.add(new_v)
                    yield Board(new_vehicles,f'{v.char} right')
                    
                    
#         def m2(self):
            
#             board = self.board
#              for car in self.cars:
                
#                 if car.axis == 'horizontal':
                    
                    
            





    def solved(self):
        """Returns true if the board is in a solved state."""
        board = self.get_board()
        return (board[2][5]=='A')
    
    
    


def breadth_first_search(r, max_depth=50):
    """
    Find solutions to given RushHour board using breadth first search.
    Returns a dictionary with named fields:
        visited: the number of configurations visited in the search
        solutions: paths to the goal state
        depth_states: the number of states visited at each depth
    Arguments:
        r: A RushHour board.
    Keyword Arguments:
        max_depth: Maximum depth to traverse in search (default=25)
    """
    visited = set()
    solutions = list()
    depth_states = dict()

    queue = deque()
    queue.appendleft((r, tuple()))
    while len(queue) != 0:
        board, path = queue.pop()
        new_path = (path + tuple([board]))

        depth_states[len(new_path)] = depth_states.get(len(new_path), 0) + 1

        if len(new_path) >= max_depth:
            break

        if board in visited:
            #print('it is')
            continue
        else:
            visited.add(board)
            print(np.array(board.get_board()))

        if board.solved():
            solutions.append(new_path)
        else:
            queue.extendleft((move, new_path) for move in board.moves())

    return {#'visited': visited,
            'solutions': solutions,
            'depth_states': depth_states}
```
