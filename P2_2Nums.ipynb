```python
class Vehicle:
    
    def __init__(self,char, axis ,x, y,length,fuel):
        if char.isalpha():
            self.char = char
        else:
            raise ValueError('Not a letter.')
        
        self.axis = axis
        
#         self.x = x
#         self.y = y
        
        if 0 <= x <=5:
            self.x = x
        else:
              raise ValueError('x is wrong size')
        if 0 <= y <=5:
            self.y = y
        else:
             raise ValueError('y is wrong size')
       
        self.length = length
            
        self.fuel = int(fuel)
        
    
    def set_fuel(val):
        self.fuel = val
    
    def set_axis(val):
        self.axis = val
    def set_x(val):
        self.x = val
    def set_y(val):
        self.y = val
    def set_length(val):
        self.length = val
        
    def __hash__(self):
        return hash((self.char,self.x,self.y))

    def __eq__(self, other):
        return self.char == other.char and self.x == other.x and self.y == other.y

    def __ne__(self, other):
        return not self.__eq__(other)

    def __repr__(self):
        return "Vehicle({0}, {1}, {2}, {3})".format(self.char, self.x, self.y,
                                                    self.axis)
    
    
            
    def __str__(self):
        return f"Vehicle({self.char},{self.axis},{self.x},{self.y},{self.length},{self.fuel})"
```

```python
 def compact(list1): 
        
    list1 = [str(x) for x in list1]
    given_list = [list(j) for i, j in groupby(list1)]
    final = []
    for i in given_list:
        if len(i) == 1:
            final.append(' '.join(i))
        else:
            i = i[0][:-1]+ str(len(i))
            final.append(i)
    

    return final
```

```python
def string_to_matrix(line,n):
    temp = [line[char: char + n] for char in range(0, len(line), n)]
    
    matrix = [list(x) for x in temp]
    
    return np.array(matrix)
    
```

```python
def matrix_vehicles(line,n):
    cars0_fuels = line.split()
    matrix = string_to_matrix(cars0_fuels[0],n)
    fuels = {}
    temp = []
    encountered = set()
    for pair in cars0_fuels[1:]:
        fuels[pair[0]] = pair[1]
    for car in cars0_fuels[0]:
        if car not in fuels:
            fuels[car] = 100
    
    for row_count,row in enumerate(matrix):
        for col,ele in enumerate(row):
            if ele == "." or ele in encountered:
                continue
            else:
                if col == len(row)-1: #end of the row / last col in row
                    if matrix[row_count][col-1] == ele: #check if left is the same value
                        axis = "horizontal"
                    else:
                        axis = "vertical"
                elif col == 0: #beginning of a row /first col in row
                    if matrix[row_count][col+1] == ele:  #check if right is the same value
                        axis = "horizontal"
                    else:
                        axis = "vertical"
                else: # somewhere in row
                    if matrix[row_count][col+1] == ele or matrix[row_count][col-1]== ele: 
                        axis = "horizontal"
                    else:
                        axis = "vertical"
                        
                encountered.add(ele)
                length = cars0_fuels[0].count(ele)
                car_fuel = fuels.get(ele)
                new_car = Vehicle(ele,axis,row_count,col,length,car_fuel)
                temp.append(new_car)
    return temp
                
        
    
```

```python
def breadth_first_search(r, max_depth=100):
   
    visited = set()
    solutions = list()
    depth_states = dict()
    g_n = dict()
    queue = deque()
    queue.appendleft((r, tuple()))
    
    while len(queue) != 0: # when we run out of new board (depleting all possible board combinations)
        board, path = queue.pop()
        #print(board.h1())
        new_path = (path + tuple([board]))
        #print(board,'--',new_path)

        depth_states[len(new_path)] = depth_states.get(len(new_path), 0) + 1

        if len(new_path) >= max_depth:
            break

        if board in visited:
            #print('old')
            continue # skips
        else: # first time we see this board
            visited.add(board)
            g_n[board]= len(new_path) -1
            #print(np.array(board.get_board()))

        if board.solved():
            solutions.append(new_path)
        else:
            #print([(move) for move in board.moves()],1)
            queue.extendleft(((move, new_path) for move in board.moves()))

    return {'visited': visited,
            'solutions': solutions,
            'depth_states': g_n}
```

```python
def GBFS1(r,max_depth=100):
    """
    Find solutions to given RushHour board using breadth first search.
    Returns a dictionary with named fields:
        visited: the number of configurations visited in the search
        solutions: paths to the goal state
        depth_states: the number of states visited at each depth
    Arguments:
        r: A RushHour board.
    Keyword Arguments:
        max_depth: Maximum depth to traverse in search (default=25)
    """
    g_n = dict()
    visited = set()
    solutions = list()
    depth_states = dict()
    queue = deque()
    queue.appendleft((r, tuple()))
    while len(queue) != 0: # when we run out of new board (depleting all possible board combinations)
        board, path = queue.pop()
        new_path = (path + tuple([board]))

        depth_states[len(new_path)] = depth_states.get(len(new_path), 0) + 1

        if len(new_path) >= max_depth:
            break

        if board in visited:
            continue # skips
        else: # first time we see this board
            visited.add(board)
            g_n[board] = len(new_path)-1
            
        if board.solved():
            solutions.append(new_path)
        else:
           
            xxx = [(move, new_path) for move in board.moves()]
            #print(len([key for key, _group in groupby(new_path)]))
            #print([new_path for move in board.moves()])
            queue.extendleft(sorted(xxx,key = lambda x:h1(x[0])))

    return {'visited': visited,
            'solutions': solutions,
            'depth_states': g_n}
```

```python
def sol_file(output,func,stuff,line):
    with open (output,'w') as f:
        cars = sorted(matrix_vehicles(stuff,6), key = lambda x: x.char)
        b = Board(set(cars),"start")
        start = perf_counter()
        result = func(b)
        stop = perf_counter()

        if(len(result['solutions']) == 0):
            f.write('No solutions found \n')
           
        else:
            
            f.write(f"Initial Board Configuration: {line}")
            f.write("\n")
            f.write("\n")
            
            f.write(string_to_board(stuff))
            f.write("\n")
            f.write("\n")

            f.write(f"Car fuel available:{[f'{x.char}: {x.fuel}' for x in cars]} \n ")
            f.write("\n")
            f.write("\n")
            
            f.write(f"Runtime: {round(stop - start,1)} seconds")
            f.write("\n")
            f.write(f"Search Path Length: {len(result['visited'])} states")
            f.write("\n")

            f.write(f"Solution Path Length: {len(goodform(result['solutions'])[1])-1} moves") # here
            f.write("\n")
            f.write(f"Solution Path :{(goodform(result['solutions']))[1][1:]} \n")
            f.write("\n")
            for moves in (goodform(result['solutions']))[0][1:]:
                f.write(f"{moves}         {[x.fuel for x in moves.cars if x.char == str(moves)[0]]} {moves.flat()}\n")
                
            f.write("\n")
            f.write("\n")
            
           
            f.write("\n")
            f.write(string_to_board((goodform(result['solutions']))[0][-1].flat()))


        f.close()  
```
